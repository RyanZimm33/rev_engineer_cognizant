```python
#String Challenge
raw_username = "Py_Thon-ROCKS123!!! "
formatted = raw_username.lstrip().lower().replace("-", "_").replace("!", "")
has_digit = "has no"
for c in formatted:
    if c.isnumeric():
        has_digit = "has a"
        break

has_python = "has" if formatted.find("python") != -1 else "does not have"
first_eight = formatted[:8]

print(f"{formatted}, String {has_digit} number and {has_python} python in it. The first eight characters are {first_eight}")

```
