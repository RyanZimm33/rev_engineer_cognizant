```python
#Challenge 1: Sort list from oldest to youngest with lambda
students = [
    ('alice', 34),
    ('bob', 22),
    ('diana', 19)
]

students.sort(key = lambda s: s[1], reverse=True)
print(students)
```

```python

#REPL app: Ask user to add an item, if yes, user types value and adds it to list and loop, if no, quit and print all items added so far.
cont = True
toPrint = []

while(cont):
    toPrint.append(input("Type line to echo: "))
    cont = input("Continue? (Y/N): ") == "Y"

print("-Result-")
for p in toPrint:
    print(p)
```

```python
#Challenge 2: Manipulate Tuples
students = [
    ("Alice", 20, "A"),
    ("Bob", 19, "B"),
    ("Charlie", 21, "A"),
    ("Diana", 20, "C")
]

part1 = [x for x in students if x[2] == "A"]
part2 = [x[0] for x in students]
part3 = [ (x[0], x[1], "B" if x[1] == 20 else x[2]) for x in students]

print(part1)
print(part2)
print(part3)
```

```python
#Challenge 3: Manipulate Sets
attendees = ["Alice", "Bob", "Alice", "Ethan", "Bob", "Diana", "Ethan", "Frank"]


attendees_set = set(attendees)

print(len(attendees_set))

if "Alice" in attendees_set:
    print("Alice is going!")

attendees_set.add("Grace")

attendees_set.discard("Bob")
attendees_set.discard("Bob")

print(attendees_set)
```

```python
#Abstract classes practice
from abc import ABC, abstractmethod

class Parent():
    @abstractmethod
    def my_method(self):
        pass

    @property
    @abstractmethod
    def id(self):
        pass

class Child1(Parent):
    def my_method(self):
        print("hi")
    
    @property
    def id(self):
        return 2


c = Child1()
print(c.id)
print(c.my_method())
    
```
